```python
text = list(encrypted_data.dropna()['tokenized'])
text
```

```python
wc = 0
for x in text:
    if x:
        wc += len(x.split())
wc
```

```python
s = set()
for x in text:
    s=s.union(set(x.split()))
```

```python
s
```

```python
len(s)
```

```python
len(text)
```

```python
s = {}

for x in text:
    tokens = x.split()
    for token in tokens:
        if token not in s:
            s[token] = 0
        s[token]+= 1
```

```python
len(s)
```

```python
encrypted_text = list(encrypted_data.dropna()['encrypted'])
encrypted_text
```

```python
se = {}

for x in encrypted_text:
    tokens = x.split()
    for token in tokens:
        if token not in se:
            se[token] = 0
        se[token]+= 1
```
